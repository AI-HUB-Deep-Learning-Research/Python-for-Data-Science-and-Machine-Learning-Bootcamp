```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
#print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# standard scaler 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
#convert 30 dimensional dataset into 3 components
pca = PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

### we just transformed our data

```python
scaled_data.shape
```

```python
x_pca.shape
```

## plotting

```python
#plotting first PCA axis VS second PCA axis
plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], cmap='coolwarm',
            edgecolors='yellow')

plt.xlabel('first principal components')
plt.ylabel('second principal components')
```

```python
#plotting second PCA axis VS third PCA axis
plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,1], x_pca[:,2], c=cancer['target'], cmap='coolwarm',
            edgecolors='yellow')

plt.xlabel('second principal components')
plt.ylabel('third principal components')
```

# Interpreting the components

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the fitted PCA object:### 

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp.corr(), cmap='plasma')
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp, cmap='plasma')
```
