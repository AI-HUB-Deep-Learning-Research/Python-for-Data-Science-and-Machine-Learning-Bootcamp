```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
tc
```

```python
sns.heatmap(tc,annot=True)
```

```python
sns.heatmap(tc, annot=True,cmap='rainbow')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='rainbow')
```

```python
sns.heatmap(fp, cmap='magma')
```

```python
sns.heatmap(fp, cmap='rainbow', linewidth=1, linecolor='white')
```

```python
sns.heatmap(fp, cmap='rainbow', linewidth=1, linecolor='black')
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp, cmap='rainbow', standard_scale=1)
```

```python
sns.jointplot(x='year',y='year',data=flights,kind='reg')
```

```python
sns.pairplot(flights)
```

```python
sns.pairplot(flights, palette='rainbow')
```

```python
sns.rugplot(flights['year'])
```
